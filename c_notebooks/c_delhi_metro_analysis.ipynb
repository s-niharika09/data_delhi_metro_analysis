# Importing Pandas and Reading JSON File

```python
import pandas as pd

df = pd.read_json(r"D:\delhi_metro_analysis\a_source_data\a_delhi_metro_trips.json")
print(df.head(10))
print("The json dataset has been successfully loaded")
```

# Revoming Extra Spaces in Station_Names using .Strip() method and Standardizing Station_Names

```python
df["From_Station"] = df["From_Station"].str.strip().str.replace(r"\s+", " ", regex= True).str.upper()
print(df["From_Station"])
df["To_Station"] = df["To_Station"].str.strip().str.replace(r"\s+", " ", regex= True).str.upper()
print(df["To_Station"])
df.head()
```

# Handling Missing Values 

```python
print(df["Ticket_Type"].isnull().sum())
print(df["Ticket_Type"].unique())
df["Ticket_Type"] = df["Ticket_Type"].fillna(df["Ticket_Type"].mode()[0])

print(df["Passengers"].isnull().sum())
print(df["Passengers"].unique())
df["Passengers"] = df["Passengers"].fillna(df["Passengers"].median())

print(df["Remarks"].isnull().sum())
print(df["Remarks"].unique())
df["Remarks"] = df["Remarks"].fillna(df["Remarks"].mode()[0])
```

# Converting Date columns into DateTime Format

```python
df["Date"].dtypes
df.info()
df.head()
# There is no requirement to convert because Date column is in DateTime Format
```

# Changing Column Name

```python
df.rename({"Date":"Jounery_Date"},axis=1,inplace=True)
```

# Ensuring Numeric Columns are Correctly Formatted

```python
# Converting Passengers from Float dtype to Int dtype
df["Passengers"] = df["Passengers"].round().astype(int)
df.head()
```

# Route Analysis

1.Which metro routes have the highest passenger traffic?

```python
import pandas as pd

df = pd.read_sql_query("""
SELECT 
    From_Station,
    To_Station,
    SUM(Passengers) as total_passengers
FROM metro_trips
GROUP BY From_Station, To_Station
ORDER BY total_passengers DESC
LIMIT 5;    
""", conn)

print(df)
```

2.Which routes generate the highest total revenue?

```python
import pandas as pd

df = pd.read_sql_query("""
SELECT
    From_Station,
    To_Station,
    Sum(Fare * Passengers) as total_revenue
FROM metro_trips
GROUP BY From_Station, To_Station
ORDER BY total_revenue DESC
LIMIT 5;
""", conn)

print(df)
```

3.What is the average fare for each route?

```python
import pandas as pd

df = pd.read_sql_query("""
SELECT
    From_Station,
    To_Station,
    AVG(Fare) as average_fare
FROM metro_trips
GROUP BY From_Station, To_Station
ORDER BY average_fare DESC
LIMIT 5;
""", conn)

print(df)
```

4.Which routes have the longest travel distances?

```python
import pandas as pd

df = pd.read_sql_query("""
SELECT
    From_Station,
    To_Station,
    MAX(Distance_km) as maximum_distance
FROM metro_trips
GROUP BY From_Station, To_Station
ORDER BY maximum_distance DESC
LIMIT 5;
""", conn)

print(df)
```

5.a.Total departure analysis:

```python

import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="sakiley",
    database="delhi_metro"
)  

cursor = conn.cursor()


import pandas as pd

df = pd.read_sql_query("""
SELECT
    station,
    SUM(departures) as total_departures,
    SUM(arrivals) as total_arrivals
FROM(
    SELECT From_Station as station, COUNT(*) as departures, 0 as arrivals
    FROM metro_trips
    GROUP BY From_Station
                       
UNION
    
    SELECT To_Station as station, 0 as arrivals, COUNT(*) as arrrival
    FROM metro_trips
    GROUP BY To_Station
)AS combined
GROUP BY station
ORDER BY total_departures DESC
LIMIT 5;
""", conn)

print(df)
```

7. What are the top 10 most frequently used metro stations?

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="sakiley",
    database="delhi_metro"
)  

cursor = conn.cursor()

import pandas as pd

df = pd.read_sql_query("""
SELECT 
    station,
    SUM(total_passengers) AS total_usage
FROM (
    SELECT From_Station AS station, SUM(Passengers) AS total_passengers
    FROM metro_trips
    GROUP BY From_Station

    UNION ALL

    SELECT To_Station AS station, SUM(Passengers) AS total_passengers
    FROM metro_trips
    GROUP BY To_Station
) AS combined
GROUP BY station
ORDER BY total_usage DESC
LIMIT 10;
""", conn)

print(df)
```
